```python
import gseapy as gp
import pandas as pd
import numpy as np
```

```python
run1 = pd.read_csv('./results/RF_gene_rank_run1.tsv', sep='\t', index_col=0)
run2 = pd.read_csv('./results/RF_gene_rank_run2.tsv', sep='\t', index_col=0)
```

```python
#UNBALANCED DISEASES
pre_res_run1 = gp.prerank(rnk=run1, gene_sets='GO_Biological_Process_2018',
                     processes=-1,
                     permutation_num=100, # reduce number to speed up testing
                     outdir=None, format='png')
```

```python
pre_res_run1.res2d.sort_values(by='fdr')
```

```python
#BALANCED DISEASES

pre_res_run2 = gp.prerank(rnk=run2, gene_sets='KEGG_2019_Human',
                     processes=-1,
                     permutation_num=100, # reduce number to speed up testing
                     outdir=None, format='png')
```

```python
pre_res_run2.res2d.sort_values(by='pval')
```
